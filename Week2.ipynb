## Question 3: View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline.

```python
discipline = data.loc[:,['Team','Yellow Cards', 'Red Cards']]
discipline
```

## Question 4: Sort the teams by Red Cards, then to Yellow Cards.

```python
discipline = discipline.sort_values(by=['Red Cards', 'Yellow Cards'],ascending=False)
discipline
```

## Question 5: Calculate the mean Yellow Cards given per Team.

```python
discipline['Yellow Cards Mean'] = (discipline['Yellow Cards']/discipline['Yellow Cards'].sum())
discipline
```

## Question 6: Filter teams that scored more than 6 goals.

```python
data.loc[data["Goals"]>5,["Team","Goals"]].sort_values("Goals",ascending=False)
```
